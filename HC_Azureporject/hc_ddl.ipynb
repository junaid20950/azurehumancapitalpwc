```python
%sql
CREATE TABLE IF NOT EXISTS hcdatabrickspwc.configs.metadata_config (
    -- Primary Key
    source_name STRING NOT NULL
        COMMENT 'Unique source identifier (e.g., employees, departments)',
    
    -- Source Configuration
    source_type STRING NOT NULL
        COMMENT 'Source type: file, table, api',
    
    landed_path STRING NOT NULL
        COMMENT 'Full path to source data',
    
    format STRING NOT NULL
        COMMENT 'File format: csv, json, parquet, tsv, avro',
    
    read_options STRING
        COMMENT 'Reader options as JSON or key=value pairs',
    
    -- Target Configuration
    bronze_table STRING NOT NULL
        COMMENT 'Target bronze table name',
    
    -- Execution Metadata
    pipeline_name STRING
        COMMENT 'Pipeline identifier',
    
    batch_id STRING
        COMMENT 'Batch identifier',
    
    run_id STRING
        COMMENT 'Run identifier',
    
    -- Schema & System
    schema_version STRING
        COMMENT 'Schema version',
    
    producer_system STRING
        COMMENT 'Source system name',
    
    ingestion_user STRING
        COMMENT 'User triggering ingestion',
    
    -- Control Flags
    is_active BOOLEAN
        COMMENT 'Whether source is enabled',
    
    priority INT
        COMMENT 'Processing order (lower = higher priority)',
    
    -- Audit Columns
    created_at TIMESTAMP
        COMMENT 'Record creation timestamp',
    
    created_by STRING
        COMMENT 'User who created this config',
    
    updated_at TIMESTAMP
        COMMENT 'Last update timestamp',
    
    updated_by STRING
        COMMENT 'User who last updated',
    
    -- Documentation
    notes STRING
        COMMENT 'Additional notes or documentation',
    
    -- Primary Key Constraint
    CONSTRAINT pk_source_name PRIMARY KEY (source_name)
)
USING DELTA
LOCATION 'abfss://configs@hcstorageaccountpwc.dfs.core.windows.net/metadata_config/'
COMMENT 'Metadata-driven ingestion configuration for Bronze layer - External table in Azure ADLS Gen2'
TBLPROPERTIES (
    'delta.autoOptimize.optimizeWrite' = 'true',
    'delta.autoOptimize.autoCompact' = 'true',
    'delta.parquet.compression.codec' = 'snappy',
    'delta.enableChangeDataFeed' = 'true',
    'delta.columnMapping.mode' = 'name'
);
```
